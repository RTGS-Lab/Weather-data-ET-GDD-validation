```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Define the locations and corresponding dataframes
locations = [
    ("Becker22", df_BeckerMDA_2022, df_GEMS_Becker2022),
    ("Clarissa22", df_ClarissaMDA_2022, df_GEMS_Clarissa2022),
    ("Hubbard22", df_HubbardMDA_2022, df_GEMS_Hubbard2022),
    ("Dakota22", df_HastingsMDA_2022, df_GEMS_Dakota2022),
    ("LittleFalls22", df_LittleFallsMDA_2022, df_GEMS_LittleFalls2022),
    ("Ottertail22", df_OtterTailMDA_2022, df_GEMS_OtterTail2022),
    ("ParkersPrairie22", df_ParkersPrairieMDA_2022, df_GEMS_ParkersPrairie2022),
    ("Perham22", df_PerhamMDA_2022, df_GEMS_Perham2022),
    ("PinePoint22", df_PinePointMDA_2022, df_GEMS_PinePoint2022),
    ("Rice22", df_RiceMDA_2022, df_GEMS_Rice2022),
    ("Staples22", df_StaplesMDA_2022, df_GEMS_Staples2022),
    ("Wadena22", df_WadenaMDA_2022, df_GEMS_Wadena2022),
    ("Westport22", df_WestportMDA_2022, df_GEMS_Westport2022)
]
# Create subplots
num_plots = len(locations)
num_cols = 3
num_rows = (num_plots + num_cols - 1) // num_cols
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))
axs = axs.flatten()

# Loop through each location
for i, (location_name, df_MDA, df_GEMS) in enumerate(locations):
    # Concatenate MDA and GEMS data after converting GEMS data to the same units as MDA
    df_combined = pd.concat([df_MDA['SolarRad_daily_mj'], df_GEMS['solar_rad']* 0.0864], axis=1)
    df_combined.columns = ['MDA', 'GEMS']
    
    # Drop NaN values
    df_combined.dropna(inplace=True)
    
    # Linear regression
    X = df_combined['MDA'].values.reshape(-1, 1)
    y = df_combined['GEMS'].values
    model = LinearRegression()
    model.fit(X, y)
    predictions = model.predict(X)

    # Calculate metrics
    r2 = r2_score(y, predictions)
    rmse = np.sqrt(mean_squared_error(y, predictions))
    mae = mean_absolute_error(y, predictions)
    
    # Plot
    axs[i].scatter(df_combined['MDA'], df_combined['GEMS'], color='gray', label='Data points', s=20)
    axs[i].plot(df_combined['MDA'], predictions, color='red', label='Regression Line', linewidth=2)
    axs[i].plot(df_combined['MDA'], df_combined['MDA'], linestyle='--', color='black', label='1:1 Line', linewidth=1)  # 1:1 Line
    axs[i].set_title(f'{location_name.replace("22", "22")}', fontsize=12)
    axs[i].legend(fontsize=10, loc='upper left', frameon=False) # Remove legend box
    axs[i].set_xlabel('MDA Solar Rad')
    axs[i].set_ylabel('GEMS Solar rad')

    # Add R2, RMSE, and MAE to the plot
    textstr = f'R-squared: {r2:.5f}\nRMSE: {rmse:.5f} mj/m^2\nMAE: {mae:.5f} mj/m^2'
    axs[i].text(0.95, 0.05, textstr, transform=axs[i].transAxes, fontsize=10,
                verticalalignment='bottom', horizontalalignment='right')

    # Adjust axis ticks
    axs[i].tick_params(axis='both', which='major', labelsize=10)
    axs[i].tick_params(axis='both', which='minor', labelsize=8)
    
    # Set y-axis limits to match x-axis limits
    x_min, x_max = axs[i].get_xlim()
    y_min, y_max = axs[i].get_ylim()
    min_limit = min(x_min, y_min)
    max_limit = max(x_max, y_max)
    axs[i].set_xlim(min_limit, max_limit)
    axs[i].set_ylim(min_limit, max_limit)

# Hide empty subplots
for i in range(num_plots, num_rows * num_cols):
    axs[i].axis('off')

# Add common title
plt.suptitle('Solar rad Comparison Across Locations 2023: GEMS (Gridded) vs. MDA (Non-Gridded)', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.99])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from scipy.stats import pearsonr

# Define the locations and corresponding dataframes
locations = [
    ("Becker22", df_BeckerMDA_2022, df_GEMS_Becker2022),
    ("Clarissa22", df_ClarissaMDA_2022, df_GEMS_Clarissa2022),
    ("Hubbard22", df_HubbardMDA_2022, df_GEMS_Hubbard2022),
    ("Dakota22", df_HastingsMDA_2022, df_GEMS_Dakota2022),
    ("LittleFalls22", df_LittleFallsMDA_2022, df_GEMS_LittleFalls2022),
    ("Ottertail22", df_OtterTailMDA_2022, df_GEMS_OtterTail2022),
    ("ParkersPrairie22", df_ParkersPrairieMDA_2022, df_GEMS_ParkersPrairie2022),
    ("Perham22", df_PerhamMDA_2022, df_GEMS_Perham2022),
    ("PinePoint22", df_PinePointMDA_2022, df_GEMS_PinePoint2022),
    ("Rice22", df_RiceMDA_2022, df_GEMS_Rice2022),
    ("Staples22", df_StaplesMDA_2022, df_GEMS_Staples2022),
    ("Wadena22", df_WadenaMDA_2022, df_GEMS_Wadena2022),
    ("Westport22", df_WestportMDA_2022, df_GEMS_Westport2022)
]

# Combine all data for regression analysis
all_data = pd.DataFrame()

for location_name, df_MDA, df_GEMS in locations:
    combined = pd.DataFrame({
        'MDA': df_MDA['SolarRad_daily_mj'],  
        'GEMS': df_GEMS['solar_rad'] * 0.0864
    })
    all_data = pd.concat([all_data, combined])

# Drop any rows with NaN values to ensure consistent lengths
all_data.dropna(inplace=True)

# Extract X and y for regression
X = all_data['MDA'].values.reshape(-1, 1)
y = all_data['GEMS'].values

# Linear regression
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# Calculate overall metrics
r2 = r2_score(y, predictions)
rmse = np.sqrt(mean_squared_error(y, predictions))
mae = mean_absolute_error(y, predictions)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='gray', label='Data points')
plt.plot(X, predictions, color='red', label='Regression Line')
plt.plot(X, X, linestyle='--', color='black', label='1:1 Line')  # 1:1 Line
plt.title('Solar radiation Combined Regression Analysis for All Locations for 2022')
plt.xlabel('MDA Solar Rad (MJ/m^2)')
plt.ylabel('GEMS Solar Rad (MJ/m^2)')

plt.legend(loc='upper left')  # Move the legend to top left

# Display R2, RMSE, and MAE
textstr = f'R-squared: {r2:.5f}\nRMSE: {rmse:.5f} MJ/m^2\nMAE: {mae:.5f} MJ/m^2'
plt.text(0.95, 0.05, textstr, transform=plt.gca().transAxes, fontsize=12,
          verticalalignment='bottom', horizontalalignment='right')

plt.tight_layout()
plt.show()

# Calculate Pearson correlation coefficient and p-value
#pearson_corr, p_value = pearsonr(all_data['MDA'], all_data['GEMS'])

# Print Pearson correlation coefficient
#print("Pearson correlation coefficient:", pearson_corr)
# Print p-value with 20 decimal places
#print("P-value:", "{:.20f}".format(p_value))
```
